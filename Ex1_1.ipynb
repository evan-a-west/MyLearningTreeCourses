23. ‚ùè Import the *requests* library, and then import the *BeautifulSoup* library from *bs4*

````python
import requests as req
from bs4 import BeautifulSoup
````

24. ‚ùè Load the URL http://en.wikipedia.org/wiki/List_of_countries_by_population into Python and view it.

````python
link = "http://en.wikipedia.org/wiki/List_of_countries_by_population"
response = req.get(link)
````

````python

links = pd.DataFrame()
soupify = BeautifulSoup(response.content.decode(), 'html.parser')
print(soupify)
````

25. ‚ùè Retrieve links from the website.

````python
links = soupify.find_all('a', href=True)

urls = []
for link in links:
    urls.append(link.get('href'))

print(urls)
````
